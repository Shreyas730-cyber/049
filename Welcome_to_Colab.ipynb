```
import pandas as pd
import matplotlib.pyplot as plt
df=pd.read_csv('/content/mtcars.csv')
auto=(df['am']==0).sum()
manual=(df['am']==1).sum()
plt.bar(['auto','manual'],[auto,manual] ,color=['red','green'])
plt.xlabel('frequency')
plt.ylabel('value')
plt.title('frequency of transmission')
plt.grid(axis='y')
plt.show()
```
